```python

import numpy as np
import matplotlib.pyplot as plt
print(np.__version__)
```

```python
fig = plt.figure()
fig.show()
```

```python
plt.plot([0,1],[0,3],label='line',color='green')
plt.title('Wykres')
plt.xlabel('os x')
plt.ylabel('os y')
_ = plt.legend()
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x,x_sin)
plt.plot(x,x_cos)
```

```python
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[2,4,1],'r--')
```

```python
fig = plt.figure(figsize=(12,8))
plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[2,4,1],'r--')
plt.subplot(133)
plt.plot([0,1,2],[7,6,5],'ro')
```

```python
plt.bar(x=[0,1,2,3],height=[1.2,3.2,1.3,5.0])
```

```python
plt.bar(x=['a','b','c','d'],height=[1.2,3.2,1.3,5.0])
```

```python
plt.barh(y=['a','b','c','d'],width=[1.2,3.2,1.3,5.0])
```
