```python
import numpy as np
import pandas as pd
print(np.__version__)
print(pd.__version__)
```

```python
def fetch_data(company='AMZN'):
  """pobiera dane spolek z serwisu stooq"""
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')
```

```python
df = fetch_data('META')
```

```python
df.head()
```

```python
df.to_csv('fb.csv')
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2023)]
```

```python
df_nov
```

```python
df_nov.to_csv('fb_nov.csv')
```
