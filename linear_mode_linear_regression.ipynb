```python
import numpy as np
```

```python
x = np.arange(0,50,0.5)
noise = 10*np.random.randn(100)
```

```python
y = 2*x+100+noise
x = x.reshape(-1,1)
```

```python
print('rozmiar x',x.shape)
print('rozmiar y',y.shape)
```

```python
x
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(x,y)
```

```python
plt.scatter(x_train,y_train, c='b',label='zbior treningowy')
plt.scatter(x_test,y_test, c='g',label='zbior testowy')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
plt.figure(figsize=(7,5))
plt.title('regresja liniowa 1 zmienna')
plt.scatter(x_train,y_train, c='b',label='zbior treningowy')
plt.plot(x,lin_reg.intercept_+lin_reg.coef_[0]*x, c='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
lin_reg.score(x_test,y_test)
```
