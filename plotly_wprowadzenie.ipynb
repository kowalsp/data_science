```python
import plotly
plotly.__version__
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(
    data=go.Bar(y=[2,3,1])
)
```

```python
fig.show()
```

```python
fig = go.Figure(
    data=go.Bar(y=[2,3,1,4]),
    layout= go.Layout(title_text='Przykladowy wykres slupkowy')

)
fig.write_html('bar.html')
fig.show()
```

```python
from seaborn import load_dataset
df = load_dataset('diamonds')
```

```python
df.head()
```

```python
dff = df.cut.value_counts()
dff = dff.reset_index()
dff.columns=['type','count']
dff
```

```python
fig = go.Figure(
    data=go.Pie(labels=dff['type'], values=dff['count'],hole=0.3),
    layout=go.Layout(title='Wykres',width=800)
)
fig.show()
```
